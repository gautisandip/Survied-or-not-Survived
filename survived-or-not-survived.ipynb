```python
train = pd.read_csv('/kaggle/input/titanic/train.csv')
```

```python
test = pd.read_csv('/kaggle/input/titanic/test.csv')
```

```python
del_col = ['Cabin','Ticket','Name']
train_data = train.drop(labels = del_col,axis = 1)
test_data = test.drop(labels = del_col, axis =1)
```

```python
train_data['Age'].fillna((train_data['Age'].mean()),inplace = True)
test_data['Age'].fillna((test_data['Age'].mean()),inplace = True)
test_data['Fare'].fillna((test_data['Fare'].mean()),inplace = True)
```

```python
train_data.isnull().sum()
```

```python
test_data.isnull().sum()
```

```python
train_data.head()
```

```python
train_data = pd.get_dummies(train_data, columns = ['Embarked','Sex'], prefix = '',prefix_sep = '')
```

```python
test_data = pd.get_dummies(test_data, columns = ['Embarked','Sex'], prefix = '',prefix_sep = '')
```

```python
test_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
predictors = train_data.drop(['Survived','PassengerId'], axis = 1)
target = train_data['Survived']
x_train,x_val,y_train,y_val = train_test_split(predictors,target,test_size = 0.22, random_state = 0)
```

```python
#Logistic regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_val)
```

```python
print(list(zip(y_val,y_pred)))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix = confusion_matrix(y_val,y_pred)
print(confusion_matrix)
```

```python
accuracy_score = accuracy_score(y_val,y_pred)
print('accuracy of model is', accuracy_score)
```
